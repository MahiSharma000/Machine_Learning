```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/data.csv')
data.drop_duplicates(inplace = True)
new_data = data.dropna()
plt.boxplot(new_data,vert = False)
plt.show()
```

```python
data = pd.read_csv("data.csv")
data.boxplot(column = ['Pulse'])
plt.show()
```

```python
data.boxplot(by='Duration', column=['Calories'],grid = False)
plt.ylabel('Calories')
plt.xlabel('Duration')
plt.show()
```
